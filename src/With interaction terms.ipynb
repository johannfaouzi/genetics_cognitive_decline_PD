# Meta-analysis

### GRS

```python
df_meta_grs = {}

for study in df_iceberg.index:

    string = (
        f"""library(meta)
        cohort <- c("ICEBERG", "LCC", "PDBP", "PPMI", "Sure")
        smd <- c({df_iceberg.loc[study, ('Effect size', 'GRS')]},
                 {df_lcc.loc[study, ('Effect size', 'GRS')]},
                 {df_pdbp.loc[study, ('Effect size', 'GRS')]},
                 {df_ppmi.loc[study, ('Effect size', 'GRS')]},
                 {df_sure.loc[study, ('Effect size', 'GRS')]})
        lower <- c({df_iceberg.loc[study, ('Lower bound (95%-CI)', 'GRS')]},
                   {df_lcc.loc[study, ('Lower bound (95%-CI)', 'GRS')]},
                   {df_pdbp.loc[study, ('Lower bound (95%-CI)', 'GRS')]},
                   {df_ppmi.loc[study, ('Lower bound (95%-CI)', 'GRS')]},
                   {df_sure.loc[study, ('Lower bound (95%-CI)', 'GRS')]})
        upper <- c({df_iceberg.loc[study, ('Upper bound (95%-CI)', 'GRS')]},
                   {df_lcc.loc[study, ('Upper bound (95%-CI)', 'GRS')]},
                   {df_pdbp.loc[study, ('Upper bound (95%-CI)', 'GRS')]},
                   {df_ppmi.loc[study, ('Upper bound (95%-CI)', 'GRS')]},
                   {df_sure.loc[study, ('Upper bound (95%-CI)', 'GRS')]})

        df <- data.frame(cohort, smd, lower, upper)

        metagen(TE = smd,
                lower = lower,
                upper = upper,
                studlab = cohort,
                sm = "dsdssdsd",
                backtransf = FALSE,
                null.effect = 0,
                data = df)"""
    )

    res_study = robjects.r(string)
    dico_study = dict(res_study.items())

    df_meta_grs[study] = {}
    df_meta_grs[study]['Coefficient (95% CI)'] = "{:.4f} [{:.4f} - {:.4f}]".format(
        dico_study['TE.fixed'][0],
        dico_study['lower.fixed'][0],
        dico_study['upper.fixed'][0]
    )
    df_meta_grs[study]['p-value (coefficient)'] = dico_study['pval.fixed'][0]
    df_meta_grs[study]['Heterogeneity p-value'] = dico_study['pval.Q'][0]

df_meta_grs = pd.DataFrame.from_dict(df_meta_grs, orient='index')
```

```python
df_meta_grs.nsmallest(10, 'p-value (coefficient)')
```

```python
df_meta_grs[df_meta_grs['p-value (coefficient)'] < 0.0001]
```

```python
df_meta_grs.to_csv(
    'Revision/Results/With interaction terms/Meta_analysis_MoCA_with_interaction_terms_GRS.tab', sep='\t'
)
```
