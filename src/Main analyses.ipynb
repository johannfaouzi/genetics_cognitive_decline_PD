# Meta-analysis

```python
df_meta = {}

for study in df_iceberg.index:

    string = (
        f"""library(meta)
        cohort <- c("ICEBERG", "LCC", "PDBP", "PPMI", "Sure")
        smd <- c({df_iceberg.loc[study, 'Coefficient']},
                 {df_lcc.loc[study, 'Coefficient']},
                 {df_pdbp.loc[study, 'Coefficient']},
                 {df_ppmi.loc[study, 'Coefficient']},
                 {df_sure.loc[study, 'Coefficient']})
        lower <- c({df_iceberg.loc[study, 'Lower bound (95% CI)']},
                   {df_lcc.loc[study, 'Lower bound (95% CI)']},
                   {df_pdbp.loc[study, 'Lower bound (95% CI)']},
                   {df_ppmi.loc[study, 'Lower bound (95% CI)']},
                   {df_sure.loc[study, 'Lower bound (95% CI)']})
        upper <- c({df_iceberg.loc[study, 'Upper bound (95% CI)']},
                   {df_lcc.loc[study, 'Upper bound (95% CI)']},
                   {df_pdbp.loc[study, 'Upper bound (95% CI)']},
                   {df_ppmi.loc[study, 'Upper bound (95% CI)']},
                   {df_sure.loc[study, 'Upper bound (95% CI)']})

        df <- data.frame(cohort, smd, lower, upper)

        metagen(TE = smd,
                lower = lower,
                upper = upper,
                studlab = cohort,
                sm = "dsdssdsd",
                backtransf = FALSE,
                null.effect = 0,
                data = df)"""
    )

    res_study = robjects.r(string)
    dico_study = dict(res_study.items())

    df_meta[study] = {}
    df_meta[study]['Coefficient (95% CI)'] = "{:.4f} [{:.4f} - {:.4f}]".format(
        dico_study['TE.fixed'][0],
        dico_study['lower.fixed'][0],
        dico_study['upper.fixed'][0]
    )
    df_meta[study]['p-value (coefficient)'] = dico_study['pval.fixed'][0]
    df_meta[study]['Heterogeneity p-value'] = dico_study['pval.Q'][0]

df_meta = pd.DataFrame.from_dict(df_meta, orient='index')
```

```python
df_meta.nsmallest(10, 'p-value (coefficient)')
```
